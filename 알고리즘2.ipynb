```python
from typing import Any, Sequence

def max_of(a: Sequence) -> Any:
    maximum =a[0]
    for i in range(1, len(a)):
        if a[i] > maximum:
            maximum = a[i]
    return maximum

if __name__ == '__main__':
    print('배열의 최댓값을 구함')
    num = int(input('원소 수 입력: '))
    x= [None] * num

    for i in range(num):
        x[i] = int(input(f'x[{i}]값 입력: '))

    print(f'{max_of(x)}')
```

```python
from typing import Any, MutableSequence

def reverse_array(a:MutableSequence) -> None:
    n = len(a)
    for i in range(n // 2):
        a[i], a[n - i - 1] = a[n - i - 1], a[i]

if __name__ == '__main__':
    print('배열을 역순으로 정렬')
    nx = int(input('원소 수 입력: '))
    x = [None] * nx

    for i in range(nx):
        x[i] = int(input(f'x[{i}]값 입력: '))

    reverse_array(x)

    for i in range(nx):
        print(f'x[{i}] = {x[i]}')
```

```python
def card_conv(x: int, r: int) -> str:
    d = ''
    dchar = '0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ'

    while x > 0:
        d += dchar[x % r]
        x //= r
    return d[::-1]

if __name__ == '__main__':
    print('10진수를 n진수로 변환')
    while True:
        while True:
            no = int(input('변환할 값으로 음이아닌 정수 입력: '))
            if no > 0:
                break

        while True:
            cd = int(input('몇 진수로 변환할지 입력: '))
            if 2 <= cd <= 36:
                break

        print(f'{cd}진수로는 {card_conv(no, cd)}')

        retry = input('한번더? (Y / N): ')
        if retry in {'N', 'n'}:
            break
```

```python
counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1

for n in range(3, 1001, 2):
    for i in range(1, ptr):
        counter += 1
        if n % prime[i] == 0:
            break
    else:
        prime[ptr] = n
        ptr += 1

print(prime[:ptr])
print(counter)

```

```python
counter = 0
ptr = 0
prime = [None] * 500

prime[ptr] = 2
ptr += 1

prime[ptr] = 3
ptr += 1

for n in range(5, 1001, 2):
    i = 1
    while prime[i] * prime[i] <= n:
        counter += 2
        if n % prime[i] == 0:
            break
        i += 1
    else:
        prime[ptr] = n
        ptr += 1
        counter += 1

print(prime[:ptr])
print(counter)

```
