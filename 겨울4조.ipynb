```python
import pandas as pd

df1 = pd.read_csv("rkdqnr.csv")

df2 = pd.read_csv("2017.csv")



print(df1)
```

```python
df1.columns
df1.drop('동별(1)', axis = 1, inplace=True)
df1.columns = ['시군구명', '읍면동명', '계 (명)', '65세이상고령자 (명)', '독거노인 (명)']
df2.drop(['코드','시도명'], axis = 1, inplace=True)
df2.sort_values(['시군구명','읍면동명'])
```

```python
merged_df = pd.merge(df1, df2, on=['시군구명', '읍면동명'], how='outer')  # 'key' 열을 기준으로 inner join 수행
merged_df
```

```python
merged_df.to_csv('merged_df.csv', index=False, encoding="utf-8-sig")
df3 = pd.read_csv('merged_df.csv')
df3
```
