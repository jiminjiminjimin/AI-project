실습 1-2

```python
def max3(a, b, c):
    maximum = a
    if b > maximum: maximum = b
    if c > maximum: maximum = c
    return maximum

print(f'max3(3, 2, 1) = {max(3, 2, 1)}')
print(f'max3(3, 2, 2) = {max(3, 2, 2)}')
print(f'max3(3, 1, 2) = {max(3, 1, 2)}')
print(f'max3(3, 2, 3) = {max(3, 2, 3)}')
```

```python
def med3(a, b, c):
    if a >= b:
        if b >= c:
            return b
        elif a <= c:
            return a
        else:
            return c
    elif a > c:
        return a
    elif b < c:
        return b
    else:
        return c

print('세 정수의 중앙값을 구합니다.')
a = int(input('정수 a의 값을 입력해 주세요.: '))
b = int(input('정수 b의 값을 입력해 주세요.: '))
c = int(input('정수 c의 값을 입력해 주세요.: '))

print(f'중앙값은 {med3(a, b, c)}입니다.')
```

```python
print('1부터 n까지 정수의 합을 구합니다.')
n = int(input('n 값을 입력하세요.'))

sum = 0
i = 1

while i <= n:
    sum += i
    i += 1

print(f'1부터 {n}까지 정수의 합은 {sum} 입니다.')
print(f'i 값은{i}')
```

실습 1-23

```python
print('오른쪽 아래가 직각인 이등변 삼각형을 출력합니다.')
n = int(input('짧은 변의 길이를 입력하세요.: '))

for i in range(n):
    for j in range(n - i - 1):
        print(' ', end = '')
    for j in range(i + 1):
        print('*', end = '')
    print()
```
