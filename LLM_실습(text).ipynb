#### ChatGPT에게 역할 부여하기

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import AIMessage, HumanMessage, SystemMessage
from langchain.callbacks.streaming_stdout import StreamingStdOutCallbackHandler

chatgpt = ChatOpenAI(model_name="gpt-4o-mini", temperature = 0.6, streaming=True, callbacks=[StreamingStdOutCallbackHandler()])

messages = [
    SystemMessage(
        content="너는 IoT전문가야. 요청한 디바이스를 만들기 위한 개발 계획서를 단계별로 작성해줘."
    ),
    HumanMessage(
        content="아두이노를 통해 시각장애인용 지팡이를 만들고 싶어."
    ),
]
response_langchain = chatgpt.invoke(messages)
response_langchain
```

### System & Human 메세지 연습해보기

```python
# 음악추천 솔루션 만들기
# 음식레시피 제조 방법 만들기

from langchain_openai import ChatOpenAI
from langchain.schema import HumanMessage, SystemMessage, AIMessage

chat = ChatOpenAI(model_name = "gpt-4o-mini", temperature = 1,
                  streaming=True, callbacks=[StreamingStdOutCallbackHandler()])
answer = chat(
  [
    SystemMessage(content="""너는 심리치료 전문가이며 음악 전문가야. 현재 상황에 맞는 음악을 추천을 해줘."""),
    HumanMessage(content="""나는 지금 수능 100일을 앞두고 있어. 너무 긴장돼""")
  ]
)

print(answer.content)
```

```python
#ChatGPT의 대답은 AIMessage라는 딕셔너리 형태로 제공됩니다.
answer
```

```python
#content를 key값으로 호출하여 를 통해 text만 출력할 수 있습니다.
print(answer.content)
```
