```python
import base64

def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')
```

```python
from langchain_openai import ChatOpenAI
llm= ChatOpenAI(model_name="gpt-4o-mini", temperature = 0.2)
```

```python
from langchain.chat_models import ChatOpenAI
from langchain.schema import AIMessage, HumanMessage, SystemMessage

image = encode_image("./fire_2.jpg")

messages =[

    SystemMessage(content="""
    너는 산불을 감시하는 로봇이야. 산불 감시에 대한 가능성을 이미지를 보고 판단해줘.
    산불 가능성을 0~4단계로 표현해줘

    ## 답변 형태
    산불 가능성 : 0단계
    이유 :
    """),

    HumanMessage(content=[
                {"type": "text",
                 "text": "이미지를 분석해줘"},
                {
                    "type": "image_url",
                    "image_url": {"url": f"data:image/jpeg;base64,{image}"},
                },
            ]
        )
    ]

result = llm.invoke(messages)
result.content
```

```python
print(result.content)
```

```python
from jikko.jikko import *
jk = Pyjikko()

# 포트 번호 알맞게 바꾸기
PORT = 'COM5'
NEO = 7

jk.serial_connect(PORT)
jk.start()

jk.neopixel_set(NEO, 4)

jk.neopixel_display_all(NEO,NeoColor.GREEN)
```

```python
if "4단계" in result.content:
    jk.neopixel_display_all(NEO,NeoColor.RED)
elif "3단계" in result.content:
    jk.neopixel_display_all(NEO,NeoColor.ORANGE)
elif "2단계" in result.content:
    jk.neopixel_display_all(NEO,NeoColor.YELLOW)
elif "1단계" in result.content:
    jk.neopixel_display_all(NEO,NeoColor.BLUE)
elif "0단계" in result.content:
    jk.neopixel_display_all(NEO,NeoColor.GREEN)
```
