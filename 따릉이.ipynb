```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
train.drop([934], axis=0, inplace=True)
train.drop([1035], axis=0, inplace=True)
```

```python
train[train['hour_bef_windspeed'].isna()]
```

```python
test[test['hour_bef_temperature'].isna()]
```

```python
import statsmodels.formula.api as smf
model1 = smf.ols('hour_bef_ozone ~ hour_bef_temperature', data=train)
results1 = model1.fit()
predictions1 = results1.predict(train['hour_bef_temperature'])
print(predictions1)
```

```python
model2 = smf.ols('hour_bef_ozone ~ hour_bef_temperature', data=test)
results2 = model2.fit()
predictions2 = results2.predict(test['hour_bef_temperature'])
print(predictions2)
```

```python
train['hour_bef_ozone']=predictions1
train['hour_bef_windspeed'].fillna(train['hour_bef_windspeed'].mean(), inplace=True)
train["hour_bef_pm10"].fillna(train["hour_bef_pm10"].mean(), inplace=True)
train["hour_bef_pm2.5"].fillna(train["hour_bef_pm2.5"].mean(), inplace=True)
```

```python
train.isnull().sum()
```

```python
test['hour_bef_ozone']=predictions2
test['hour_bef_windspeed'].fillna(test['hour_bef_windspeed'].mean(), inplace=True)
test["hour_bef_pm10"].fillna(test["hour_bef_pm10"].mean(),inplace=True)
test["hour_bef_pm2.5"].fillna(test["hour_bef_pm2.5"].mean(), inplace=True)
test = test.fillna(test.mean())
```

```python
test.isnull().sum()
```

```python
train_x = train.drop(['count'],axis = 1)
train_y = train['count']
```

```python
model=RandomForestRegressor(n_estimators=100)
model.fit(train_x,train_y)
```

```python
pred = model.predict(test)
```

```python
submission = pd.read_csv('submission.csv')
submission
```

```python
submission['count'] = pred
submission
```
