```python
# Selenium 설정
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
options = webdriver.ChromeOptions()
options.add_argument('--headless')  # Headless 모드
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')
options.add_argument('--window-size=1920x1080')
options.add_argument('--remote-debugging-port=9222')  # 디버깅 포트 설정


import time
url = "https://www.instagram.com"
driver = webdriver.Chrome('chromedriver', options=options)
driver.get(url)

# 로그인하기
userid = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="loginForm"]/div/div[1]/div/label/input')))
userid.send_keys("temp_id1027")
userpw = WebDriverWait(driver, 5).until(EC.presence_of_element_located((By.XPATH, '//*[@id="loginForm"]/div/div[2]/div/label/input')))
userpw.send_keys("d7fXUip0DyMGS8o")
login_button = WebDriverWait(driver, 5).until(EC.element_to_be_clickable((By.XPATH, '//*[@id="loginForm"]/div/div[3]/button')))
login_button.click()
time.sleep(5)

# 로그인 정보 저장 "나중에 하기" 누르기 위해
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
div = soup.find('div', string='나중에 하기')
class_list = div.get('class')
class_name = class_list[0]
login_infor = driver.find_element(By.CLASS_NAME, class_name) # WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, class_name))) 동일
login_infor.click()
time.sleep(5)

# 알림 설정 "나중에 하기" 누르기 위해
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
target_div = soup.find('button', string='나중에 하기')
class_list = target_div.get('class')
class_name = class_list[0]
alert_button = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CLASS_NAME, class_name)))
alert_button.click()
```

```python
# 계정 들어가기
account1 = "eol.essenceoflove"

svg_element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'svg[aria-label="검색"]')))
svg_element.click()

search_element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, 'input[aria-label="입력 검색"]')))

search_element.send_keys(account1)
time.sleep(5)

xpath = f"//span[text()='{account1}']"
span_element = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, xpath)))
span_element.click()
```

```python
# 페이지 내리면서 ("div", "_aagv")인 애들 모두 가져와서 img 링크만 파싱 후에 저장
page_height = driver.execute_script("return document.body.scrollHeight") # 맨 처음 page_height 가져오기
print("first page height: ",page_height)

all_aagv_list = []

while True:
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);") # 문서의 body의 scrollheight만큼 스크롤하기
    time.sleep(2)
    new_page_height = driver.execute_script("return document.body.scrollHeight")
    if page_height == new_page_height: # 페이지 높이가 달라지지 않았다면 끝까지 스크롤 한 것 !
        break;
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")
    aagv_list = soup.find_all("div", "_aagv")
    for aagv in aagv_list:
        if aagv not in all_aagv_list:  # 중복되지 않으면 추가
            all_aagv_list.append(aagv)
    page_height = new_page_height
```

```python
# eol.essenceoflove 게시물 개수: 417? 근데 몇 개 빠진지도.. ㅠㅠㅠ
# len(all_aagv_list)
```

```python
# 그 다음에 전체 img 링크만 파싱 후에 저장
for i, aagv in enumerate(all_aagv_list):
    img_html = aagv.find('img')
    img_src = img_html["src"]
    response = requests.get(img_src)
    file_name = f"./img/{account1}_{i}_img.jpg"
    if response.status_code == 200:
        with open(file_name, 'wb') as file:
            file.write(response.content)
```

```python
import openai
import time

# OpenAI API 키 설정
openai.api_key = 'your-api-key-here'

def generate_text(prompt):
    while True:
        try:
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo",  # 또는 원하는 모델
                messages=[{"role": "user", "content": prompt}]
            )
            return response.choices[0].message['content']
        except openai.error.RateLimitError:
            print("Rate limit exceeded. Waiting before retrying...")
            time.sleep(60)  # 1분 대기 후 재시도

# 예제 프롬프트
prompt = "인스타그램 피드에서 사용할 재미있는 문구를 만들어주세요."
generated_text = generate_text(prompt)
print(f"Generated Text: {generated_text}")
```
